* 예시 데이터
    * 날짜와 도시 단위의 패널 데이터 구조
    * t는 시간을, D는 처치 변수를 지칭
    * 마케팅 팀은 $D_i = 1$ 인 도시에서 오프랑니 캠페인 진행
    * 처치변수 D는 $t > T_{pre}$ 일 때 1 이고 그렇지 않으면 0인 더미변수$
    * 개입은 개입 후 기간인 $t > T_{pre}$ 의 실험군에만 발생
    * 실제 처치 효과 : $\tau$
* 패널 데이터에서의 ATT 수식
    * $ATT = E[Y_{it}(1) - Y_{it}(0)|D=1,t > T_{pre}]$
    * 위 ATT 는 캠페인 시작후 캠페인이 D=1 인 도시들에 캠페인이 미친 영향을 이해하는데만 초점을 맞춘 식이다.

* 관측된 결과를 아래 그림과 같이 실험 대상 - 시간 행렬로 표현할 때, 개입 후 기간의 실험군에서만 Y(1) 을 관측할 수 있고 그 외 3개에 셀에서는 Y(0)을 관측할 수 있다.
* 누락된 잠재적 결과를 구하기 위해 개입 후 기간의 대조군 결과를 사용해 실험 대상 간의 상관관계를 활용할 수 있고, 개입 전 기간의 실험군 결과를 상요해 시간 간 상관관계를 활용할 수도 있다.
* 즉, 실험군에 대한 반사실 Y(0)을 패널데이터 에서 추정하기 훨씬 쉽다는 의미이다.

## 8.2 표준 이중차분법
* 이중차분법(DID)의 기본 개념은 관측된 실험군 기준값에 대조군 결과 추세를 보정해 누락된 잠재적 결과인 $E[Y(0)|D=1,Post=1]$ 를 추정하는 것이다.
    * $E[Y(0)|D=1, Post=1] = \underbrace{E[Y|D=1, Post=0]}_{실험군 기준값} + \underbrace{(E[Y|D=0,Post=1]-E[Y|D=0,Post=0])}_{대조군 결과 추세}$
* $E[Y(0)|D=1,Post=1]$ 를 추정할 때, 앞의 수식에서 오른쪽 기댓값을 표번평균으로 대체할 수 있다.
* $E[Y(0)|D=1,Post=1]$ 를 ATT에 대입하면, 말 그대로 차이의 차이를 구할 수 있기 때문에 DID라고 부르는 것이다.
    * $ATT = (E[Y|D=1, Post =1]-E[Y|D=1, Post=0]) - (E[Y|D=0, Post =1]-E[Y|D=0, Post=0])$
* 먼저 데이터의 기간을 개입 전과 후로 나눈다. -> 실험 대상을 실험군과 대조군으로 구분한다. -> 마지막으로 네 개의 셀 모두의 평균을 계산한다.

### 8.2.1 이중차분법과 결과 변화
* 실험 대상 i에 대한  시간 경과에 따른 결과의 차이를 $\vartriangle y_t = E[y_i|t >T_{pre}] - E[y_i|t \leq T_{pre}]$ 로 정의할 때
* 잠재적 결과 표기법으로 ATT 를 정의하면 아래와 같다.
    * $ATT = E[\vartriangle y_1 - \vartriangle y_0]$
    * 이중차분법은 $\vartriangle y_0$ 를 대조군 평균으로 대체해 ATT 를 식별한다.
        * $ATT = E[\vartriangle y | D=1] - E[\vartriangle y | D=0]$

```python
pre = mkt_data.query("post==0").groupby("city")["downloads"].mean()
post = mkt_data.query("post==1").groupby("city")["downloads"].mean()

delta_y = ((post - pre)
           .rename("delta_y")
           .to_frame()
           # add the treatment dummy
           .join(mkt_data.groupby("city")["treated"].max()))

delta_y.tail()
```

```python
(delta_y.query("treated==1")["delta_y"].mean() 
 - delta_y.query("treated==0")["delta_y"].mean())
```

* 아래 그래프는 이중차분법을 시간에 따라 관측된 실험군 및 대조군 결과와 추정된 실험군의 반사실적 결과를 시각화 한 결과이다.
* $E[Y(0)|D=1]$ 의 DID는 점섬으로 표시된다. 이는 대조군 추세를 실험군 기준값에 보정하여 얻은 결과이다. 
* 추정된 ATT는 처치 후 기간에 추정된 반사실적 결과 $Y(0)$ 과 관측된 결과 $Y(1)$ 의 차이이다.

### 8.2.2 이중차분법과 OLS
* 기간 별 도시 데이터셋을 사용해 아래의 선형 모델을 추정할 수 있다.
    * $Y_{it} = \beta_0 + \beta_1D_i + \beta_2Post_i + \beta_3D_iPost_i + e_{it}$
    * 회귀계수 $\hat{\beta_3}$ 는 DID 추정값이다.
    * $\beta_0$ 는 대조군 기준값이며 2021-05-15 이전 기간의 다운로드 수준이다.
    * $\beta_0 + \beta_1$ 은 개입전의 실험군 도시 기준값이며
    * $\beta_1$ 은 단순히 실험군 도시와 대조군 도시간 기준값 차이이다. >> 개입에 따른 효과를 측정하기 위해 실험군과 대조군의 차이를 설명
    * $\beta_0 + \beta_2$ 는 개입 후 대조군 도시의 다운로드 수준이다.
    * $\beta_2$ 는 대조군의 추세이며 ,개입 전후로 대조군이 얼마나 변했는지를 설명한다. >> 시간이 지남에 따라 발생하는 개입 전후의 변화량을 설명
    * $\beta_0 + \beta_1 + \beta_2 + \beta3$ 는 개입 후 실험군 도시의 다운로드 수준
    * 정리하면, $\beta_3$ 는 실험군과 대조군간 차이와 개입 전후의 시간 추세를 모두 고려한 추정량

### 8.2.3 이중차분법과 고정효과
* 이중차분법을 보는 또 다른 관점 중 하나는 고정효과모델(fixed-effect model) 즉, 이원고정효과를 사용하는 것이다.
* 수식은 아래와 같은데 이 모델에서는 처치효과, 개별 대상과 시간 고정효과가 존재한다.
    * $Y_{it} = \tau W_{it}+\alpha_i+\gamma_i + e_{it} \\\\ W_{it} = D_i Post_t$
    * $\tau$ 는 처치효과 , 개발 대상에 대한 효과는 $\alpha_i$, 시간 고정효과는 $\gamma_i$ 이다.
* 여기서 W의 매개변수 추정값이 앞서 얻은 DID 추정값과 일치하며 ATT를 구할 수 있다.
* (상호작용 항 하나만 원할 때는 : 연산자를 사용한다.)

```python
# 코드 상 수식에서 교호 작용 항 하나만 넣을 경우 콜론 : 을 활용한다.
m = smf.ols('downloads ~ treated:post + C(city) + C(post)',
            data=did_data).fit()

m.params["treated:post"]
```

### 8.2.5 추론
* 패널데이터에서 신뢰구간 개념을 포함한 추론은 굉장히 까다롭다.
* 문제는 N*T 개의 데이터 포인트가 있지만 동일한 실험대상이 여러번 나타나 이 데이터가 독립적이고 동일하게 분포(iid; independent and identically distributed) 되지 않았다는 점이다.
* 회귀 분석에서 지나치게 낙관적인 표준오차를 보정하기 위해 이를 실험 대상 별로 군집화하면 된다. 

```python
m = smf.ols(
    'downloads ~ treated:post + C(city) + C(date)', data=mkt_data
).fit(cov_type='cluster', cov_kwds={'groups': mkt_data['city']})
print(m.summary())
print("ATT:", m.params["treated:post"])
m.conf_int().loc["treated:post"]
```

* 위에서 구한 군집 표준오차가 기준 표준오차보다 신뢰구간이 더 넓은 것을 볼 수 있다.

* 일별 데이터프레임 구조인 데이터를 실험대상과 전후 기간별로 집계한 데이터 (did_data)로 대체할 때 신뢰구간이 넓어지는 것도 확인할 수 있다.
    * 표본 크기가 시간대가 아닌 실험 대상에 의해 결정되어야 한다는 걸 감안하더라도, 각 군집당 더 많은 시간대가 있으면 분산을 줄일 수 있음

```python
m = smf.ols('downloads ~ treated:post + C(city) + C(date)',
            data=mkt_data).fit()

print("ATT:", m.params["treated:post"])
m.conf_int().loc["treated:post"]
```

* 전체 추정 과정을 부트스트랩하는 방법을 통해 신뢰구간을 구할 수 잇고, 이는 반복되는 실험 대상이 있으므로 동일 대상에 대한 오차가 상관될 것이라는 점에 주의해야 한다. 따라서 전체 실험 대상으로 복원추출을 시행해야한다. (블록 부트스트랩)
    * 실험 대상을 복원추출하는 함수를 정의한다.
    * 전체 데이터에 적용된 DID 추정량의 95%신뢰구간을 계산한다.

```python
def block_sample(df, unit_col):
    
    units = df[unit_col].unique()
    sample = np.random.choice(units, size=len(units), replace=True) 
    
    return (df
            .set_index(unit_col)
            .loc[sample]
            .reset_index(level=[unit_col]))
```

```python
from joblib import Parallel, delayed

def block_bootstrap(data, est_fn, unit_col,
                    rounds=200, seed=123, pcts=[2.5, 97.5]):
    np.random.seed(seed)
    
    stats = Parallel(n_jobs=4)(
        delayed(est_fn)(block_sample(data, unit_col=unit_col))
        for _ in range(rounds))
    
    return np.percentile(stats, pcts)
```

```python
def est_fn(df):
    m = smf.ols('downloads ~ treated:post + C(city) + C(date)',
                data=df).fit()
    return m.params["treated:post"]

block_bootstrap(mkt_data, est_fn, "city")
```

## 8.3 식별 가정


### 8.3.1 평행 추세
* DiD 는 평행 추세를 가정한다.
    * 이중차분법은 처치가 없으면 평균적으로 실험군과 대조군의 결과 추세가 동일할 것이라는 상대적으로 약한 가정을 깔고 있다.
    * $E[Y(0)_{it=1} - Y(0)_{it=0}|D=1] = E[Y(0)_{it=1}-Y(0)_{it=0}|D=0]$
* 평행추세 가정은 관측할 수 없는 항 (반사실) $E[Y(0)_{it}=1 |D=1]$ 이 포함되므로 검증할 수 없다.

```python
def did_date(df, date):
    df_date = (df
               .query("date==@date | post==0")
               .query("date <= @date")
               .assign(post = lambda d: (d["date"]==date).astype(int)))
    
    m = smf.ols(
        'downloads ~ I(treated*post) + C(city) + C(date)', data=df_date
    ).fit(cov_type='cluster', cov_kwds={'groups': df_date['city']})
    
    att = m.params["I(treated * post)"]
    ci = m.conf_int().loc["I(treated * post)"]
    
    return pd.DataFrame({"att": att, "ci_low": ci[0], "ci_up": ci[1]},
                        index=[date])
```

```python
post_dates = sorted(mkt_data["date"].unique())[1:]

atts = pd.concat([did_date(mkt_data, date)
                  for date in post_dates])

atts.head()
```

```python
cohorts = sorted(mkt_data_cohorts_w["cohort"].unique())

treated_G = cohorts[:-1]
nvr_treated = cohorts[-1]

def did_g_vs_nvr_treated(df: pd.DataFrame,
                         cohort: str,
                         nvr_treated: str,
                         cohort_col: str = "cohort",
                         date_col: str = "date",
                         y_col: str = "downloads"):
    did_g = (
        df
        .loc[lambda d:(d[cohort_col] == cohort)|
                      (d[cohort_col] == nvr_treated)]
        .assign(treated = lambda d: (d[cohort_col] == cohort)*1)
        .assign(post = lambda d:(pd.to_datetime(d[date_col])>=cohort)*1)
    )
    
    att_g = smf.ols(f"{y_col} ~ treated*post",
                    data=did_g).fit().params["treated:post"]
    size = len(did_g.query("treated==1 & post==1"))
    return {"att_g": att_g, "size": size}


atts = pd.DataFrame(
    [did_g_vs_nvr_treated(mkt_data_cohorts_w, cohort, nvr_treated)
     for cohort in treated_G]
)
    
atts
```
